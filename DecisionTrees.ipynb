```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
iris = load_iris()
```

```python
iris.feature_names, iris.data
```

```python
iris.target, iris.target_names
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(iris.data,iris.target)
```

```python
plt.figure(figsize=(10,12))
tree.plot_tree(classifier,filled=True)
```

## Random Forest Implementation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
rf_cls = RandomForestClassifier(oob_score=True, random_state=0)

param_grid = {
'n_estimators': [100, 150],
'max_features': ['auto', 'sqrt', 2],
'max_depth': [3, 6, 9, None],
'min_samples_split': [2, 4, 6, 7, 9],
'min_samples_leaf': [1, 2, 3, 4],
"criterion": ["gini", "entropy"],
"oob_score": [True, False]
}

grid_search = GridSearchCV(estimator=rf_cls, param_grid=param_grid, cv=5)
grid_search.fit(X_train,y_train)
```

```python
print('Best parameters:',grid_search.best_params_)
print('OOB error:',grid_search.best_estimator_.oob_score_)
```
